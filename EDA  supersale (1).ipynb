```python
# import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import data

df = pd.read_csv(
    r"C:\Users\vr380\OneDrive\Documents\panda_tutorial\Task2\Sample - Superstore.csv",
    encoding="latin1"
)

```

```python
# Inspecting first 5 rows

print("First 5 rows are :")
df.head()
```

```python
#inspecting last 5 rows
print("Last five rows are:")
df.tail()
```

```python
# Defining shape (columns x rows)
print("Shape of dataset is :")
df.shape
```

```python
# Finding information about the dataset

print("Information of dataset :")
df.info()
```

```python
# datatypes

print("Different datatypes according to the columns :")
df.dtypes
```

```python
#dropping fully blank row

df = df.dropna(how ="all")
print("After dropping blank row" , df.shape)
```

```python
# Checking and handling missing values

print("Number of null values:")
df.isnull().sum()
```

```python
# Checking dataframe that has index changes to row_id

print("Dataframe is :")
df
```

```python
# finding duplicate value

print("number of duplicate values :")
dup_value = df.duplicated().sum()
dup_value
```

```python
# Dropping duplicates

print("After removing duplicates :")
drop_dup= df.drop_duplicates()
drop_dup
```

```python
# Rechecking shape if the duplicates are removed 
print("Shape changes from (9994,21) to :")
drop_dup.shape
```

```python
# changing to date column

df["Order Date"] = pd.to_datetime(df["Order Date"], format='mixed')
df["Ship Date"] = pd.to_datetime(df["Ship Date"], format='mixed')
```

```python
# changing date column with years only

df["Order Date"]=df["Order Date"].dt.year
df["Ship Date"]=df["Ship Date"].dt.year
```

```python
print("Dataframe after cleaning the data :")
df.head()
```
